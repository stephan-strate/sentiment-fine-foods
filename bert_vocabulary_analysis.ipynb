```python
from urllib.request import urlopen

import matplotlib.pyplot as plt
import matplotlib as mpl

TU_GREEN = "#639a00"

vocab = urlopen("https://raw.githubusercontent.com/stephan-strate/sentiment-fine-foods/master/data/uncased_L-12_H-768_A-12_vocab.txt").read().decode('utf-8').splitlines()
print("{:,} Tokens".format(len(vocab)))

print("Duplicates? " + str(len(set(vocab)) != len(vocab)))
```

```python
plt.title("Token length distribution")
plt.xlabel("Chars")
plt.ylabel("Absolute count")
plt.hist(list(map(lambda x: len(x), vocab)), color=TU_GREEN)
ax = plt.gca()
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
plt.savefig("images/vocab_token_length_distribution.pdf")
plt.show()
```

```python
one_char_tokens = list(filter(lambda x: len(x) == 1, vocab))
print("One char tokens: {:,}, {}".format(len(one_char_tokens), str(one_char_tokens)))
```

```python
one_char_prefix_tokens = list(map(lambda x: x[2], filter(lambda x: x.startswith("##") and len(x) == 3, vocab)))
print("One char prefix tokens: {:,}, {}".format(len(one_char_prefix_tokens), str(one_char_prefix_tokens)))
```

```python
one_char_tokens.sort()
one_char_prefix_tokens.sort()

print("All one char tokens also available with prefix? {}".format(str(one_char_tokens == one_char_prefix_tokens)))
```

```python
import pandas as pd

df_chinese_characters = pd.read_csv("https://raw.githubusercontent.com/stephan-strate/sentiment-fine-foods/master/data/hanzi_db_2020-08-29.csv", delimiter=",", encoding="utf-8")

chinese_characters = df_chinese_characters.charcter.values
print("Most common chinese characters: {:,}, {}".format(len(chinese_characters), chinese_characters))

chinese_tokens = list(filter(lambda x: x in chinese_characters, vocab))
print("Most common chinese characters in vocab: {:,}, {}".format(len(chinese_tokens), chinese_tokens))
```

```python
unused_tokens = list(filter(lambda x: "unused" in x and x.startswith("[") and x.endswith("]"), vocab))
print("Unused tokens: {:,}, {}".format(len(unused_tokens), unused_tokens))
```

```python
special_tokens = list(filter(lambda x: x.startswith("[") and x.endswith("]") and not "unused" in x, vocab))
print("Special tokens: {:,}, {}".format(len(special_tokens), special_tokens))
```

```python
prefix_tokens = list(filter(lambda x: x.startswith("##"), vocab))
print("Prefix tokens: {:,}, {}".format(len(prefix_tokens), prefix_tokens))
```

```python
from names_dataset import NameDataset

m = NameDataset()
name_tokens = list(filter(lambda x: m.search_first_name(x) or m.search_last_name(x), vocab))
print("Name tokens: {:,}, {}".format(len(name_tokens), name_tokens))
```

```python
other_tokens = list(filter(lambda x: x not in [*unused_tokens, *special_tokens, *prefix_tokens, *one_char_tokens, *name_tokens], vocab))

data = {
    "special tokens": special_tokens,
    "unused tokens": unused_tokens,
    "one char tokens": one_char_tokens,
    "name tokens": name_tokens,
    "prefix tokens": prefix_tokens,
    "others": other_tokens,
}
```

```python
plt.figure(figsize=(6, 6))
patches, texts = plt.pie(list(map(lambda x: len(x), data.values())), startangle=90)
plt.legend(patches, ["{}: {:,}".format(k, len(v)) for (k, v) in data.items()], loc="lower right")
plt.savefig("images/vocab_partitioning.pdf")
plt.show()
```
